```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

# Simulated data
temperature = np.linspace(300, 600, 30)        # K
yield_strength = 500 - 0.1*temperature + np.random.normal(0, 5, 30)

df = pd.DataFrame({
    "Temperature": temperature,
    "Yield_Strength": yield_strength
})
```

```python
from sklearn.linear_model import LinearRegression

X = df[["Temperature"]]   # input (must be 2D)
y = df["Yield_Strength"]  # output

model = LinearRegression()
model.fit(X, y)
```

```python
slope = model.coef_[0]
intercept = model.intercept_
print(f"Slope: {slope:.3f}")
print(f":Intercept:{intercept:.3f}")
```

```python
y_pred = model.predict(X)
plt.scatter(X, y)
plt.plot(X, y_pred)
plt.xlabel("Temperature (K)")
plt.ylabel("Yield Strength (MPa)")
plt.title("Linear Regression: Yield Strength vs Temperature")
plt.show()
```
